Load IBM Granite Model

```python
MODEL_NAME = "ibm-granite/granite-3.3-2b-instruct"

print("⏳ Loading IBM Granite Model... This may take a minute.")

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

# Load model
model = AutoModelForCausalLM.from_pretrained(
    MODEL_NAME,
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
    device_map="auto"
)

print("✅ IBM Granite model loaded successfully!")
```

AI HTML Generator

```python
def generate_ai_html(prompt):
    """
    Generate HTML content using IBM Granite model
    """

    # Tokenize prompt
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)

    # Generate output
    outputs = model.generate(
        **inputs,
        max_new_tokens=600,
        temperature=0.4,
        top_p=0.9,
        do_sample=True
    )

    # Decode result
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    return generated_text
```

HTML Page Builder

```python


class HTMLPageBuilder:
    """
    Responsible for building the HTML structure of the website
    """

    def __init__(self):
        self.available_sections = [
            "hero",
            "features",
            "about",
            "portfolio",
            "testimonials",
            "pricing",
            "contact",
            "footer"
        ]


    # ------------------------------------------------------
    # Detect sections based on description keywords
    # ------------------------------------------------------
    def detect_sections(self, description):

        description = description.lower()
        detected = []

        for section in self.available_sections:
            if section in description:
                detected.append(section)

        # Ensure footer always exists
        if "footer" not in detected:
            detected.append("footer")

        return detected


    # ------------------------------------------------------
    # Generate SEO Meta Tags
    # ------------------------------------------------------
    def generate_meta_tags(self, title, description):

        meta = f"""
<meta charset="UTF-8">
<meta name="viewport" content="width=device-width, initial-scale=1.0">
<meta name="description" content="{description}">
<meta name="keywords" content="website, portfolio, modern site">
<meta name="author" content="HTML Quick Styler AI">
<title>{title}</title>
"""
        return meta


    # ------------------------------------------------------
    # Generate HTML for each section
    # ------------------------------------------------------
    def generate_section_html(self, section, title):

        if section == "hero":
            return f"""
<section class="hero">
<div class="container">
<h1>Welcome to {title}</h1>
<p>We create beautiful and modern web experiences.</p>
<button class="btn-primary">Get Started</button>
</div>
</section>
"""

        elif section == "features":
            return """
<section class="features">
<div class="container">
<h2>Features</h2>

<div class="feature-grid">

<div class="card">
<h3>Fast</h3>
<p>Lightning fast performance.</p>
</div>

<div class="card">
<h3>Responsive</h3>
<p>Works on all devices.</p>
</div>

<div class="card">
<h3>Modern</h3>
<p>Clean modern design.</p>
</div>

</div>
</div>
</section>
"""

        elif section == "about":
            return """
<section class="about">
<div class="container">
<h2>About Us</h2>
<p>We are passionate about building amazing digital products and experiences.</p>
</div>
</section>
"""

        elif section == "portfolio":
            return """
<section class="portfolio">
<div class="container">
<h2>Portfolio</h2>

<div class="portfolio-grid">
<div class="card">Project 1</div>
<div class="card">Project 2</div>
<div class="card">Project 3</div>
</div>

</div>
</section>
"""

        elif section == "testimonials":
            return """
<section class="testimonials">
<div class="container">
<h2>Testimonials</h2>

<div class="card">
<p>"Amazing service and beautiful design!"</p>
</div>

</div>
</section>
"""

        elif section == "pricing":
            return """
<section class="pricing">
<div class="container">
<h2>Pricing</h2>

<div class="pricing-grid">

<div class="card">
<h3>Basic</h3>
<p>$9/month</p>
</div>

<div class="card">
<h3>Pro</h3>
<p>$29/month</p>
</div>

</div>
</div>
</section>
"""

        elif section == "contact":
            return """
<section class="contact">
<div class="container">
<h2>Contact Us</h2>

<form>
<input type="text" placeholder="Name">
<input type="email" placeholder="Email">
<textarea placeholder="Message"></textarea>
<button type="submit">Send</button>
</form>

</div>
</section>
"""

        elif section == "footer":
            return """
<footer>
<div class="container">
<p>© 2026 HTML Quick-Styler. All rights reserved.</p>
</div>
</footer>
"""

        return ""


    # ------------------------------------------------------
    # Build full HTML structure
    # ------------------------------------------------------
    def build_full_page(self, title, description):

        sections = self.detect_sections(description)

        html_sections = ""

        for sec in sections:
            html_sections += self.generate_section_html(sec, title)

        meta_tags = self.generate_meta_tags(title, description)

        html_page = f"""
<!DOCTYPE html>
<html lang="en">

<head>
{meta_tags}
<link rel="stylesheet" href="styles.css">
</head>

<body>

{html_sections}

</body>
</html>
"""

        return html_page


# Create builder instance
html_builder = HTMLPageBuilder()

print("🌐 HTML Page Builder Ready!")
```

Website Generator Pipeline

```python
# ==========================================================
# HTML Quick-Styler Project
# SECTION 7 — Website Generator Pipeline
# ==========================================================

def generate_website(title, description, palette_name):
    """
    Main pipeline that generates the full website
    """

    # 1️⃣ Get selected color palette
    palette = palette_generator.get_palette(palette_name)

    # 2️⃣ Generate CSS using palette
    css = generate_css(palette)

    # 3️⃣ Generate HTML structure using builder
    html_structure = html_builder.build_full_page(title, description)

    # 4️⃣ Generate AI content using Granite
    ai_prompt = f"""
Create modern website content for a page.

Title: {title}

Description:
{description}

Write short professional website text content for sections like
hero, features, about, testimonials and contact.
"""

    ai_content = generate_ai_html(ai_prompt)

    # 5️⃣ Combine CSS and HTML into a single document
    final_html = html_structure.replace(
        '<link rel="stylesheet" href="styles.css">',
        f"<style>\n{css}\n</style>"
    )

    # 6️⃣ Insert AI content comment (optional for future expansion)
    final_html += f"\n<!-- AI Generated Content -->\n<!-- {ai_content[:500]} -->"

    # 7️⃣ Save HTML file
    file_path = "generated_website.html"

    with open(file_path, "w", encoding="utf-8") as f:
        f.write(final_html)

    return final_html, file_path


print("🚀 Website Generator Pipeline Ready!")
```

Gradio Interface

```python
# ==========================================================
# HTML Quick-Styler Project
# SECTION 8 — Gradio Interface
# ==========================================================

import gradio as gr


# Function to handle UI generation
def ui_generate(title, description, palette):

    html, file_path = generate_website(title, description, palette)

    return html, html, file_path


# ---------------------------------------
# Palette Display Function
# ---------------------------------------

def show_palettes():

    html = ""

    for name in palette_generator.list_palettes():

        p = palette_generator.get_palette(name)

        html += f"""
        <div style="margin-bottom:20px;padding:15px;border:1px solid #ddd;border-radius:10px">
        <h3>{name.title()} Palette</h3>

        <div style="display:flex;gap:10px">

        <div style="width:40px;height:40px;background:{p['primary']}"></div>
        <div style="width:40px;height:40px;background:{p['secondary']}"></div>
        <div style="width:40px;height:40px;background:{p['accent']}"></div>
        <div style="width:40px;height:40px;background:{p['background']}"></div>
        <div style="width:40px;height:40px;background:{p['surface']}"></div>

        </div>
        </div>
        """

    return html


# ---------------------------------------
# Example Templates
# ---------------------------------------

templates_html = """
<h2>Example Website Ideas</h2>

<ul>
<li>Startup Landing Page</li>
<li>Portfolio Website</li>
<li>Restaurant Website</li>
<li>AI SaaS Product Page</li>
<li>Personal Blog</li>
<li>Photography Portfolio</li>
</ul>
"""


# ---------------------------------------
# Guide Section
# ---------------------------------------

guide_html = """
<h2>How to Use HTML Quick-Styler</h2>

<ol>
<li>Enter your website title</li>
<li>Describe the page layout</li>
<li>Select a color palette</li>
<li>Click Generate Website</li>
<li>Preview the page</li>
<li>Download the HTML file</li>
</ol>

<p>This tool generates responsive websites using AI.</p>
"""


# ==========================================================
# Build Gradio UI
# ==========================================================

with gr.Blocks(title="HTML Quick-Styler") as demo:

    gr.Markdown("# 🚀 HTML Quick-Styler")
    gr.Markdown("Generate complete modern websites using AI.")

    with gr.Tabs():

        # ====================================
        # TAB 1 — GENERATE PAGE
        # ====================================

        with gr.Tab("Generate Page"):

            title_input = gr.Textbox(label="Page Title")

            description_input = gr.Textbox(
                label="Page Description",
                lines=6
            )

            palette_dropdown = gr.Dropdown(
                palette_generator.list_palettes(),
                label="Color Palette",
                value="modern"
            )

            generate_btn = gr.Button("Generate Website")

            html_output = gr.Code(
                label="Generated HTML Code",
                language="html"
            )

            download_file = gr.File(label="Download HTML File")


        # ====================================
        # TAB 2 — LIVE PREVIEW
        # ====================================

        with gr.Tab("Live Preview"):

            preview_html = gr.HTML()


        # ====================================
        # TAB 3 — COLOR PALETTES
        # ====================================

        with gr.Tab("Color Palettes"):

            gr.HTML(show_palettes())


        # ====================================
        # TAB 4 — TEMPLATES
        # ====================================

        with gr.Tab("Templates"):

            gr.HTML(templates_html)


        # ====================================
        # TAB 5 — GUIDE
        # ====================================

        with gr.Tab("Guide"):

            gr.HTML(guide_html)


    # ====================================
    # BUTTON ACTION
    # ====================================

    generate_btn.click(
        fn=ui_generate,
        inputs=[title_input, description_input, palette_dropdown],
        outputs=[html_output, preview_html, download_file]
    )


print("🎨 Gradio UI Ready!")
```
